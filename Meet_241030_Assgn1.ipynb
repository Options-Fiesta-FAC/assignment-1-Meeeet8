```python
from matplotlib import pyplot as plt
import numpy as np
K = int(input("Enter the Strike price "))
P = int(input("Enter the Premium you paid/recieved "))
spotrng = np.linspace(K-50,K+50,100)
def buy_call(S,K,P):
    return np.maximum(0,S-K) - P
def sell_call(S,K,P):
    return -np.maximum(0,S-K) + P
def buy_put(S,K,P):
    return np.maximum(0,K-S) - P
def sell_put(S,K,P):
    return -np.maximum(0,K-S) + P
bc = buy_call(spotrng,K,P)
sc = sell_call(spotrng,K,P)
bp = buy_put(spotrng,K,P)
sp = sell_put(spotrng,K,P)
plt.plot(spotrng,bc,color = "blue")
plt.plot(spotrng,sc,color = "black")
plt.plot(spotrng,bp,color = "green")
plt.plot(spotrng,sp,color = "red")
plt.show()
```
